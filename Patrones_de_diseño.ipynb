```python
# FACTORY METHOD - Ejemplo 1: Sistema de notificaciones

class Notificacion:
    """Clase base que todas las notificaciones deben seguir"""
    def enviar(self, mensaje):
        pass

class NotificacionEmail(Notificacion):
    def enviar(self, mensaje):
        print(f"[EMAIL] Enviando mensaje: '{mensaje}'")

class NotificacionSMS(Notificacion):
    def enviar(self, mensaje):
        print(f"[SMS] Enviando mensaje: '{mensaje}'")

class NotificacionPush(Notificacion):
    def enviar(self, mensaje):
        print(f"[PUSH] Enviando notificación: '{mensaje}'")


class FabricaNotificaciones:
    """
    La fábrica recibe un tipo y devuelve el objeto correcto.
    El código principal no necesita saber cómo se crea cada notificación.
    """
    def crear_notificacion(self, tipo):
        if tipo == "email":
            return NotificacionEmail()
        elif tipo == "sms":
            return NotificacionSMS()
        elif tipo == "push":
            return NotificacionPush()
        else:
            raise ValueError(f"Tipo de notificación desconocido: {tipo}")


#PROGRAMA PRINCIPAL
print(" FACTORY METHOD: Notificaciones \n")

fabrica = FabricaNotificaciones()

tipos = ["email", "sms", "push"]

for tipo in tipos:
    notificacion = fabrica.crear_notificacion(tipo)
    notificacion.enviar("Tu pedido ha sido enviado")
```

```python
# FACTORY METHOD - Ejemplo 2: Fábrica de animales

class Animal:
    """Clase base para todos los animales"""
    def hablar(self):
        pass

    def moverse(self):
        pass

class Perro(Animal):
    def hablar(self):
        print("[PERRO] Guau guau!")

    def moverse(self):
        print("[PERRO] Corriendo con cuatro patas")

class Pajaro(Animal):
    def hablar(self):
        print("[PÁJARO] Pío pío!")

    def moverse(self):
        print("[PÁJARO] Volando con las alas")

class Pez(Animal):
    def hablar(self):
        print("[PEZ] Gluh gluh")

    def moverse(self):
        print("[PEZ] Nadando con las aletas")


class FabricaAnimales:
    """
    La fábrica centraliza la creación de animales.
    Si mañana añadimos un Gato, solo tocamos esta clase.
    El resto del código no cambia.
    """
    def crear_animal(self, tipo):
        if tipo == "perro":
            return Perro()
        elif tipo == "pajaro":
            return Pajaro()
        elif tipo == "pez":
            return Pez()
        else:
            raise ValueError(f"Animal desconocido: {tipo}")


#PROGRAMA PRINCIPAL
print(" FACTORY METHOD: Animales \n")

fabrica = FabricaAnimales()
animales = ["perro", "pajaro", "pez"]

for tipo in animales:
    print(f" Creando {tipo}")
    animal = fabrica.crear_animal(tipo)
    animal.hablar()
    animal.moverse()
    print()
```

```python
# STRATEGY - Ejemplo 1: Algoritmos de ordenación

class EstrategiaOrdenacion:
    """Clase base que todas las estrategias de ordenación deben seguir"""
    def ordenar(self, lista):
        pass

class OrdenacionBurbuja(EstrategiaOrdenacion):
    """Compara pares de elementos y los va intercambiando"""
    def ordenar(self, lista):
        lista = lista.copy()
        n = len(lista)
        for i in range(n):
            for j in range(0, n - i - 1):
                if lista[j] > lista[j + 1]:
                    lista[j], lista[j + 1] = lista[j + 1], lista[j]
        print(f"[BURBUJA]   Resultado: {lista}")
        return lista

class OrdenacionInsercion(EstrategiaOrdenacion):
    """Va colocando cada elemento en su posición correcta"""
    def ordenar(self, lista):
        lista = lista.copy()
        for i in range(1, len(lista)):
            clave = lista[i]
            j = i - 1
            while j >= 0 and lista[j] > clave:
                lista[j + 1] = lista[j]
                j -= 1
            lista[j + 1] = clave
        print(f"[INSERCIÓN] Resultado: {lista}")
        return lista

class OrdenacionPython(EstrategiaOrdenacion):
    """Usa el algoritmo interno de Python, el más eficiente"""
    def ordenar(self, lista):
        resultado = sorted(lista)
        print(f"[PYTHON]    Resultado: {resultado}")
        return resultado


class Ordenador:
    """
    El contexto. No sabe cómo ordenar, delega en la estrategia asignada.
    Puedes cambiar la estrategia en cualquier momento.
    """
    def __init__(self, estrategia: EstrategiaOrdenacion):
        self.estrategia = estrategia

    def cambiar_estrategia(self, estrategia: EstrategiaOrdenacion):
        self.estrategia = estrategia

    def ordenar(self, lista):
        return self.estrategia.ordenar(lista)


#PROGRAMA PRINCIPAL
print(" STRATEGY: Ordenación \n")

numeros = [64, 25, 12, 98, 43, 7, 55]
print(f"Lista original: {numeros}\n")

ordenador = Ordenador(OrdenacionBurbuja())
ordenador.ordenar(numeros)

ordenador.cambiar_estrategia(OrdenacionInsercion())
ordenador.ordenar(numeros)

ordenador.cambiar_estrategia(OrdenacionPython())
ordenador.ordenar(numeros)
```

```python
# STRATEGY - Ejemplo 2: Descuentos en una tienda

class EstrategiaDescuento:
    """Clase base que todas las estrategias de descuento deben seguir"""
    def aplicar(self, precio):
        pass

class SinDescuento(EstrategiaDescuento):
    def aplicar(self, precio):
        print(f"[SIN DESCUENTO]  Precio final: {precio:.2f}€")
        return precio

class DescuentoVIP(EstrategiaDescuento):
    """Los clientes VIP tienen siempre un 20% de descuento"""
    def aplicar(self, precio):
        resultado = precio * 0.80
        print(f"[VIP 20%]        {precio:.2f}€ → {resultado:.2f}€")
        return resultado

class DescuentoRebajas(EstrategiaDescuento):
    """En rebajas se aplica un 50% de descuento"""
    def aplicar(self, precio):
        resultado = precio * 0.50
        print(f"[REBAJAS 50%]    {precio:.2f}€ → {resultado:.2f}€")
        return resultado

class DescuentoCupon(EstrategiaDescuento):
    """El cupón descuenta una cantidad fija en euros"""
    def __init__(self, cantidad_fija):
        self.cantidad_fija = cantidad_fija

    def aplicar(self, precio):
        resultado = max(0, precio - self.cantidad_fija)
        print(f"[CUPÓN -{self.cantidad_fija}€]      {precio:.2f}€ → {resultado:.2f}€")
        return resultado


class Carrito:
    """
    El carrito no calcula descuentos por sí solo.
    Delega ese trabajo en la estrategia que tenga asignada.
    Cambiar el descuento es tan simple como cambiar la estrategia.
    """
    def __init__(self, estrategia: EstrategiaDescuento):
        self.estrategia = estrategia

    def cambiar_descuento(self, estrategia: EstrategiaDescuento):
        self.estrategia = estrategia

    def calcular_precio(self, precio_base):
        return self.estrategia.aplicar(precio_base)


# PROGRAMA PRINCIPAL
print(" STRATEGY: Descuentos\n")

precio_producto = 120.00
print(f"Precio base del producto: {precio_producto:.2f}€\n")

carrito = Carrito(SinDescuento())
carrito.calcular_precio(precio_producto)

carrito.cambiar_descuento(DescuentoVIP())
carrito.calcular_precio(precio_producto)

carrito.cambiar_descuento(DescuentoRebajas())
carrito.calcular_precio(precio_producto)

carrito.cambiar_descuento(DescuentoCupon(30))
carrito.calcular_precio(precio_producto)
```
